```python
import pandas as pd

data = {
    'customer_id': [1, 2, 3, 4, 5],
    'customer_name': ['Arjun', 'Sneha', 'Akash', 'Ritu', 'Meena'],
    'city': ['Pune', 'Mumbai', 'Pune', 'Nagpur', 'Mumbai'],
    'product': ['Mobile', 'Laptop', 'Headphones', 'Tablet', 'Mobile'],
    'quantity': [1, 1, 2, 1, 2],
    'price': [20000, 60000, 2000, 25000, 20000]
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv("customer_data.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("customer_data.csv")
df.head()
```

```python
df['total_purchase'] = df['quantity'] * df['price']
df.head()
```

```python
df.groupby('customer_name')['total_purchase'].sum().sort_values(ascending=False)
```

```python
df.groupby('product')['quantity'].sum().sort_values(ascending=False)
```

```python
df.groupby('city')['total_purchase'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df.groupby('product')['quantity'].sum().plot(kind='bar')
plt.title("Product Sales")
plt.show()
```

```python
df.groupby('city')['total_purchase'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("City Revenue")
plt.show()
```
